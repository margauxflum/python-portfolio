```python
import pandas as pd
df = pd.read_csv("mxmh_survey_results.csv")
```

```python
mean1 = df["Anxiety"].mean()
median1 = df["Anxiety"].median()
mean2 = df["Depression"].mean()
median2 = df["Depression"].median()
```

```python
print(f"The mean for anxiety is: {mean1}.")
print(f"The median for anxiety is: {median1}.")
print(f"The mean for depression is: {mean2}.")
print(f"The median for depression is: {median2}.")
```
